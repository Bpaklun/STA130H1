### Analysis of Data
Multiple methods of analysis can be used to clearly determine if regularly sending check-in messages to friends or family has positively association with life satisfaction and inverse association with feelings of emotional loneliness

I will firstly convert my ordinal variables into numerical values.

**Spearman Rank Correlation:**

The spearman rank of correlation determines if there is a linear correlation between two variables. Two tests can be conducted, one to find the correlation of `CONNECTION_activities_checked_in_p3m` with `WELLNESS_life_satisfaction` and then `LONELY_dejong_emotional_social_loneliness_scale_miss`. By finding Spearman's rank correlation coefficient, we can see if there is a correlation between these variables. To conduct this test, we should assume both variables should be ordinal and related monotonically.

**Bootstrapped Sampling and Sampling Distribution with a Hypothesis test**

For the first test, I will determine whether checking in raises life satisfaction. I will first separate the sample into two groups based on `CONNECTION_activities_checked_in_p3m`, where frequent check-ins will be people who checked in above the median frequency and infrequent check-ins the people who checked in below the median frequency. I will take each group's life satisfication data perform bootstrapping and find the mean for each sample. Then I will find the difference between the means between the frequent and infrequent check in groups, and calculate the 95% confidence interval for the difference in means using the bootstrapped distribution. If the 95% confidence interval does not include 0, we reject the null hypothesis, showing that there is evidence to suggest that checking in frequently positively influences life satisfaction.

Secondly, I will determine whether or not checking in makes individuals feel less lonely. I will do the same in seperating the sample into two groups (frequent and infrequent check-ins) and then create a sampling distribution from the difference of means from bootstrapping. If the 95% confidence interval does not include 0, we reject the null hypothesis, showing that there is evidence to suggest that checking in frequently causes individuals to feel less lonely.

### Hypothesis/results and discussion

If both tests are rejected, then my hypothesis will be true. If only one is rejected, it shows that one variable more or so benefits well-being, which could allow for exploration using other methods. If both are rejected, there may not be enough evidence to support the hypothesis that checking in benefits well-being. This investigation links well with the course objective, as it shows if checking in on others can lead to individuals feeling better about their life and also feel less lonely, which can benefit their well-being. If the investigation follows my hypothesis, these results can be used to raise awareness about the importance in checking in on others and finding time to engage in social interactions, which could positively affect mental health.
